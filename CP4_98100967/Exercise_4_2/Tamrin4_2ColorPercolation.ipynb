```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
from scipy.ndimage import measurements
```

```python
def rand_lattice(L: int=20, p: float=0.5):
    """
    It creates a random lattice (matrix of booleans) with size L by L.
    The elements are True with probability (p) and False with probability (q=1-p).
    """


    return np.random.rand(L,L) <= p


def is_percolating(L_by_L):
    """
    checks the percolation of a lattice
    """


    labeled_lattice, num_clusters = measurements.label(L_by_L)
    if np.size(np.intersect1d(labeled_lattice[:,0][labeled_lattice[:,0]>0], labeled_lattice[:,-1][labeled_lattice[:,-1]>0])) == 0:
        return 0
    else:
        return 1
```

```python
# Plotting the L by L image matrix with imshow():
fig, ax = plt.subplots(2, 2, figsize=(12,12))
# ax.clear()
ax[0,0].imshow(rand_lattice(), origin="lower", cmap="BuPu")
ax[0,0].set_title(f"L=20, p=0.5")
ax[0,0].set_axis_off()

ax[0,1].imshow(rand_lattice(L=20, p=0.7), origin="lower", cmap="BuPu")
ax[0,1].set_title("L=20, p=0.7")
ax[0,1].set_axis_off()

ax[1,0].imshow(rand_lattice(L=100, p=0.5), origin="lower", cmap="BuPu")
ax[1,0].set_title("L=100, p=0.5")
ax[1,0].set_axis_off()

ax[1,1].imshow(rand_lattice(L=100, p=0.7), origin="lower", cmap="BuPu")
ax[1,1].set_title("L=100, p=0.7")
ax[1,1].set_axis_off()
```
