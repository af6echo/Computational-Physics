```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use("dark_background")
```

```python
Ns = [5, 10, 100, 1000]
num_samples = 100_000

fig, ax = plt.subplots(2, 2, figsize=(12,12))
fig.suptitle("Central Limit Theorem\n\
Samples Drawn from Uniform Distribution of Integers 0 to 9; Number of Samples: 100,000")
for i in range(len(Ns)):
    ys = np.mean(np.random.randint(low=0, high=10, size=(num_samples, Ns[i])), axis=1)
    ax.ravel()[i].hist(ys, rwidth=0.8, bins=20, density=True, alpha=0.7)
    ax.ravel()[i].set_title(f"N = {Ns[i]}", fontsize=15)
    ax.ravel()[i].set_xlabel("Average")
    ax.ravel()[i].set_ylabel("Normalized Occurrence")
    ax.ravel()[i].set_xticks(list(range(10)) + [4.5])
    ax.ravel()[i].tick_params(axis="x", which="major", labelsize=8)
    # The next two lines correspond to standard normal distribution dashed line:
    x = np.linspace(0, 9, 100)
    ax.ravel()[i].plot(x, stats.norm.pdf(x, (0+9)/2, np.std(ys)), "r--")
```
