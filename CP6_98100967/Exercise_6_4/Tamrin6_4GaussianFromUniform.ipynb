```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use("dark_background")
```

```python
sample_size = 50
# Generating two sets of random numbers from Uniform Distribution in the interval [0,1):
x = np.random.rand(sample_size)
xprime = np.random.rand(sample_size)
# sigma is arbitrary, so I used the variance of Standard Normal Distribution:
sigma = 1
# Converting Uniform Dist to Normal Dist:
rho = np.sqrt(-2 * sigma**2 * np.log(1-x))
theta = 2 * np.pi * xprime
y1 = rho * np.cos(theta)
y2 = rho * np.sin(theta)
# Plotting:
fig, ax = plt.subplots()

ax.hist(y1, bins=20, rwidth=0.8, density=True, alpha=0.7)
ax.set_title(f"Standard Normal Distribution Generated from Standard Uniform Distribution\n\
Sample Size = {sample_size}", fontsize=10)
ax.set_xlabel("Number")
ax.set_ylabel("Normalized Occurrence")
x = np.linspace(-3*sigma, 3*sigma, 100)
ax.plot(x, stats.norm.pdf(x, 0, sigma), "r--");
```
