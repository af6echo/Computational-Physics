```python
import numpy as np, matplotlib.pyplot as plt
```

```python
# The very first point is set, arbitrarily, at (x,y) = (unit, unit) on the plot, as defined below:
unit = 10
point = np.array([unit,unit])
# Initialize an empty list for x's and y's of the points.
# An element will be added to these lists after each iteration.
list_of_xs = list()
list_of_ys = list()
# This many points will be generated and plotted
number_of_points = 4000
number_of_transformations = 20
# The three scaling functions as described in section 2.1 of the text book.
def f1(x):
    """
    Only scales to half. It falls on the lower left corner
    """
    return x/2


def f2(x):
    """
    Scales to half and then transitions a little to the right.
    """
    return np.array([x[0]/2 + unit/2, x[1]/2])


def f3(x):
    """
    Scales to half and transitions to right and up.
    """
    return np.array([x[0]/2 + unit/4, x[1]/2 + unit/2])


# Generate all the points
for n in range(number_of_points):
    # Apply the scaling funcions number_of_transformations=20 times: f1f2f2f1f3f3f1f2... 20 times on the point.
    # The algorithm starts with the first point (unit, unit) and the value of point
    # changes with every iteration and that is how the fractal is fabricated.
    for P in range(number_of_transformations):
        a = (np.random.randint(1,3+1))
        if a==1:
            point = f1(point)
        elif a==2:
            point=f2(point)
        elif a==3:
            point=f3(point)
    # After each iteration, the x and y of the point obtained is stored in the corresponding lists below;
    # notice that the lists do not contain the first point (unit, unit).
    list_of_xs.append(point[0])
    list_of_ys.append(point[1])


# Plotting all the points together:
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(list_of_xs, list_of_ys, marker="s")
ax.set_axis_off();
ax.set_title("The Sierpinski Triangle with Random Algorithm\n(4000 Points and 20 Iterations)");
# The line of code below emphasizes the very last point obtained:
# ax.scatter(point[0], point[1], color="r", s=100);
```
