```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

$\Large p \ = \ 0.6$

```python
p = 0.6
# number of steps of a random walker:
N = 10000
# number of random walkers generated:
num_samples = 5000
# the number of times we look at the walker and record its position:
num_snapshots = 200
# Each row of data corresponds to the snapshots taken from one walker:
data = np.zeros((num_samples, num_snapshots), dtype=int)

for i in range(num_samples):
    # Generate one random walk and let it take N steps; record its position after every step it takes:
    position = np.cumsum((2 * (np.random.random(N) < p) - 1))
    # Take snapshots from the walker and put it into a row of data:
    data[i,:] = position[range(N//num_snapshots-1, N+1, N//num_snapshots)]
```

$\Large Plotting \ for \ \langle x \rangle \ and \ \sigma^2 = \langle x^2 \rangle - \langle x \rangle^2$

```python
# This is the range for the x axis; it corresponds to time:
xs = np.arange(1, num_snapshots+1) * N//num_snapshots
position_mean = np.mean(data, axis=0)
sigma2 = (np.mean((data**2), axis=0)) - (np.mean(data, axis=0)**2)

coefs1 = np.polynomial.polynomial.polyfit(xs, position_mean, deg=1)
ffit1 = np.polynomial.polynomial.polyval(xs, coefs1)
coefs2 = np.polynomial.polynomial.polyfit(xs, sigma2, deg=1)
ffit2 = np.polynomial.polynomial.polyval(xs, coefs2)

fig, ax = plt.subplots(1,2, figsize=(12, 6))
plt.suptitle("Random Walker Position Mean and Varaince vs. Time\n10,000 steps, 5,000 samples, p=0.6")
plt.subplots_adjust(wspace=0.4)

# the plot for mean:
ax[0].scatter(xs, position_mean, color="orange")
ax[0].plot(xs, ffit1, "b--", label=f"y = {coefs1[1]:.3f}x + ({coefs1[0]:.3f});\n(p-q)=0.2")
ax[0].set_title("Mean Position")
ax[0].set_xlabel("t (time or number of steps)", fontsize=12)
ax[0].set_ylabel("$\langle x \\rangle$ (mean position)", fontsize=12)
ax[0].legend();

# the plot for variance (sigma2):
ax[1].scatter(xs, sigma2, color="orange")
ax[1].plot(xs, ffit2, "b--", label=f"y = {coefs2[1]:.3f}x + ({coefs2[0]:.3f});\n4(pq)=0.96")
ax[1].set_title("Position Variance")
ax[1].set_xlabel("t (time or number of steps)", fontsize=12)
ax[1].set_ylabel("$\sigma ^2$ (variance of position)", fontsize=12)
ax[1].legend();
```
